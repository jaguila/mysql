```python
import requests
def fetch_search_results(
    query=None, minAsk=None, maxAsk=None, bedrooms=None
):
    # creates dictionary for all the non-empty input variables above
    search_params = {
        key: val for key, val in locals().items() if val is not None
    }
    #create error out
#     if not search_params:
#         raise ValueError("No valid keywords")

    # base = 'http://seattle.craigslist.org/search/apa'
    base="http://sfbay.craigslist.org/search/apa"

    # run the requests
    # opener = AppURLopener()
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:20.0) Gecko/20100101 Firefox/20.0'}
    resp = requests.get(base, params=search_params, timeout=3, headers=headers)
    #resp = requests.get(base, headers=headers)
    # resp = opener.open(base)

    return resp
```
